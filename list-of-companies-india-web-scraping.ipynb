```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Step 1: URL of wikipedia
url = "https://en.wikipedia.org/wiki/List_of_companies_of_India"
# Step 2: Add headers to behave like a real browser
headers = {

    "user-Agent" : "Mozilla/5.0"
}
# Step 3: Send request
response = requests.get(url, headers = headers)
content = response.text

# Step 4: Convert HTML into BeautifulSoup object
soup = BeautifulSoup(content,'html.parser')

# Step 5: Find all company cards

table_heading = soup.find_all("th", rowspan="2")

# extract columns of the table
col_name =[]
for i in table_heading:
    a = i.get_text(strip= True)
    col_name.append(a)

# Extract table rows
table_row = soup.find_all("tr",style="background:#ffffff;")

all_companies = []

for i in table_row:
    table_data = i.find_all("td")
    # Create dictionary using zip
    one_row = {}
    for colname, tablevalue in zip(col_name,table_data):
        one_row[colname] = tablevalue.get_text(strip=True, separator = " ")

# # Extract company link
    linkcol = i.find("a")
    one_row["Go_To"] = f"https://en.wikipedia.org/{linkcol.get("href"," ")}"
    
    all_companies.append(one_row)



df = pd.DataFrame(all_companies)

df = df.rename(columns={"Headquarters" : "HQ City","Founded": "Founded Year","Notes" : "Root Business"})

df.head(20)

df.to_excel(r'C:\Users\hp\Downloads\scraping\List_of_companies.xlsx', sheet_name= 'Details',index= False)

print("File saved successfully!")
```
